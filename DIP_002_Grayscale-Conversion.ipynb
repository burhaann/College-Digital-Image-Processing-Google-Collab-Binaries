```python
import cv2
import matplotlib.pyplot as plt

# Mount Google Drive to access the image
from google.colab import drive
drive.mount('/content/drive')

# File path of the image in Google Drive
image_path = '/content/drive/MyDrive/Colab Notebooks/DIP-DATASETS/burhaan.jpg'  # Replace with the path to your image file

# Load the image in grayscale
original_image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

# Get the image dimensions
height, width = original_image.shape

# Number of sections to divide the image into (change this value as needed)
n = 3

# Calculate the size of each section
section_height = height // n
section_width = width // n

# Extract and display each section separately
section_images = []
for i in range(n):
    for j in range(n):
        section = original_image[i*section_height:(i+1)*section_height, j*section_width:(j+1)*section_width]
        section_images.append(section)

# Plot the original image in grayscale
plt.figure(figsize=(40, 40))
plt.subplot(n+1, n, 1)
plt.imshow(original_image, cmap='gray')
plt.title('Original Image')

# Plot each section
plt.figure(figsize=(10, 8))
for i in range(n*n):
    plt.subplot(n+1, n, n + i + 1)
    plt.imshow(section_images[i], cmap='gray')
    # plt.title(f'Section {i+1}')
    plt.axis('off')

plt.tight_layout()
plt.show()
```
