```python
# 001. Printing first 100 natural Numbers and saving them in a list.
numbers = []
print("First 100 Natural Numbers")
for i in range(1, 101):
  numbers.append(i)
print(numbers)

# Print last 50 Elements of that list in both Forward and Reverse order.
print("\nLast 50 Natural Numbers of the above List")
print(numbers[-50:])

reverse = []
print("\n50 Ending Elements in Reverse Order")
for i in range(len(numbers),50,-1):
  reverse.append(i)
print(reverse)
print("\nLength of Reverse List = " + str(len(reverse)) + "\n")
```

```python
# 003. Graphs and PieCharts using Diabetes Dataset.

import pandas as pd

# Loading the Video Game Sales dataset into a Pandas DataFrame from google drive
from google.colab import drive
drive.mount('/content/drive',force_remount=True)
#data = pd.read_csv("/content/datasets/DIP_003_diabetes.csv")
folder = "Colab Notebooks/DIP-DATASETS"
file_path = f"/content/drive/MyDrive/{folder}/DIP_003_diabetes.csv"
data = pd.read_csv(file_path)

import matplotlib.pyplot as plt
import seaborn as sns

#Distribution of Age in Indian Diabetes Dataset
sns.set(style='whitegrid')
plt.figure(figsize=(10, 6))
sns.histplot(data['Age'], bins=20, kde=True)
plt.xlabel('Age', fontsize=14)
plt.ylabel('Count', fontsize=14)
plt.title('Distribution of Age in Indian Diabetes Dataset', fontsize=16)
plt.show()

#Percentage of Diabetes in Indian Diabetes Dataset
labels = ['No Diabetes', 'Diabetes']
sizes = data['Outcome'].value_counts()
colors = ['lightgreen', 'lightcoral']
plt.figure(figsize=(6, 6))
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=140)
plt.axis('equal')
plt.title('Percentage of Diabetes in Indian Diabetes Dataset', fontsize=16)
plt.show()
```

```python
#Creating a Triangle (Stright and Upside Down), Parallelogram, Hexagon and a Star Shape

import matplotlib.pyplot as plt
import numpy as np

# Create a straight triangle
def straight_triangle():
    triangle_points = np.array([[0, 0], [0.5, np.sqrt(3)/2], [1, 0], [0, 0]])
    plt.plot(triangle_points[:, 0], triangle_points[:, 1])

# Create an upside-down triangle
def upside_down_triangle():
    triangle_points = np.array([[0, np.sqrt(3)/2], [0.5, 0], [1, np.sqrt(3)/2], [0, np.sqrt(3)/2]])
    plt.plot(triangle_points[:, 0], triangle_points[:, 1])

# Create a parallelogram
def parallelogram():
    parallelogram_points = np.array([[0, 0], [0.5, np.sqrt(3)/2], [1.5, np.sqrt(3)/2], [1, 0], [0, 0]])
    plt.plot(parallelogram_points[:, 0], parallelogram_points[:, 1])

# Create a hexagon
def hexagon():
    hexagon_points = np.array([[0, 0], [0.5, np.sqrt(3)/2], [1.5, np.sqrt(3)/2],
                               [2, 0], [1.5, -np.sqrt(3)/2], [0.5, -np.sqrt(3)/2], [0, 0]])
    plt.plot(hexagon_points[:, 0], hexagon_points[:, 1])

# Create a star shape
def star():
    star_points = np.array([[0, 1], [-0.588, -0.809], [0.951, 0.309],
                            [-0.951, 0.309], [0.588, -0.809], [0, 1]])
    plt.plot(star_points[:, 0], star_points[:, 1])

# Set up the figure
plt.figure(figsize=(6, 6))

# Draw the shapes
plt.subplot(321)
straight_triangle()
plt.title('Straight Triangle')

plt.subplot(322)
upside_down_triangle()
plt.title('Upside-Down Triangle')

plt.subplot(323)
parallelogram()
plt.title('Parallelogram')

plt.subplot(324)
hexagon()
plt.title('Hexagon')

plt.subplot(325)
star()
plt.title('Star Shape')

# Adjust layout and display the plot
plt.tight_layout()
plt.show()
```
